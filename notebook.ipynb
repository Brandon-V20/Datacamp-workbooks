```python
import random

for i in range(10):
    dice1 = random.randint(1,6)
    dice2 = random.randint(1,6)
    total = dice1 + dice2
    
    if total == 7 or total == 11:
        print("A win!")
    elif total == 2 or total == 3 or total == 12:
        print("A loss!")
    else:
        print("Roll again!")
```
